```python


file_path = "E:/Javna_parkiralista_za_bicikle.csv"

# Load the CSV file with coordinates
data = pd.read_csv(file_path)

# Print column names to see what columns are available
print(data.columns)
```

```python


# Rename columns if needed
data = data.rename(columns={
    'X': 'Geografska dužina',
    'Y': 'Geografska širina',
    'OBJECTID': 'Object ID',
    'vrsta': 'Vrsta',
    'naziv': 'Ime',
    'lokacija': 'Lokacija',
    'broj_stalaka': 'Broj stalaka',
    'broj_bicikala': 'Broj bicikala',
    'godina_postavljanja':'Godina postavljanja',
    'MJESNI_ODBOR':'Mjesni odbor',
    'GRADSKA_CETVRT':'Gradska četvrt'
})

# Create a map centered at a specific location
m = folium.Map(location=[data['Geografska širina'].mean(), data['Geografska dužina'].mean()], zoom_start=10)

# Plot each coordinate on the map
for index, row in data.iterrows():
    popup_text = "<h2>Podaci o lokaciji</h2><ul>"
    for col, val in row.iteritems():
        popup_text += f"<li><strong>{col}:</strong> {val}</li>"
    popup_text += "</ul>"
    
    iframe = folium.IFrame(html=popup_text, width=400, height=200)
    popup = folium.Popup(iframe, max_width=2650)
    
    folium.Marker([row['Geografska širina'], row['Geografska dužina']], popup=popup).add_to(m)

# Show the map
m.save('map.html')  # Save the map as an HTML file
```
